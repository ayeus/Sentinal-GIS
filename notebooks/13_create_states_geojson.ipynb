```python
import geopandas as gpd
from pathlib import Path

# Load your existing state-year GeoJSON
gdf = gpd.read_file("../data/processed/india_dengue_state_year.geojson")

# Keep only one geometry per state
states_gdf = gdf[["state", "geometry"]].drop_duplicates(subset="state")

# Ensure CRS
states_gdf = states_gdf.to_crs(epsg=4326)

# Create backend data folder if not exists
output_path = Path("../backend/data")
output_path.mkdir(parents=True, exist_ok=True)

# Save clean state boundaries GeoJSON
states_gdf.to_file(
    output_path / "india_states.geojson",
    driver="GeoJSON"
)

print("india_states.geojson created successfully.")
```
