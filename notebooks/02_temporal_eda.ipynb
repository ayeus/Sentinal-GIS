```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

gdf = gpd.read_file("../data/processed/india_dengue_state_year.geojson")

# Aggregate national cases per year
yearly_cases = gdf.groupby("year")["cases"].sum().reset_index()

plt.figure()
plt.plot(yearly_cases["year"], yearly_cases["cases"], marker='o')
plt.xlabel("Year")
plt.ylabel("Total Dengue Cases")
plt.title("National Dengue Trend Over Time")
plt.grid(True)
plt.show()
```
